```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
# load data
data = pd.read_csv("../data/biomechanical_classification.csv")
```

```python
# find features data with some correlation
sns.heatmap(data.corr(), annot=True)
plt.show()
```

```python
# data = data.loc[data["class"] == "Abnormal"]
x = np.array(data.loc[:, "pelvic_incidence"]).reshape(-1, 1)
y = np.array(data.loc[:, "sacral_slope"]).reshape(-1, 1)
```

```python
plt.scatter(x=x, y=y)
plt.xlabel("pelvic_incidence")
plt.ylabel("sacral_slope")
plt.show()
```

```python
# fit model, predict and score
reg = LinearRegression()

reg.fit(x,y)
predictions = reg.predict(x)

print("R^2 score:", reg.score(x, y))
```

```python
plt.plot(x, predictions, color='black', linewidth=3)
plt.scatter(x=x,y=y)
plt.xlabel('pelvic_incidence')
plt.ylabel('sacral_slope')
plt.show()
```

```python
# cross validation
cv_result = cross_val_score(reg, x, y, cv=5)
print('CV Scores: ', cv_result)
print('CV scores average: ', np.sum(cv_result) / 5)
```
