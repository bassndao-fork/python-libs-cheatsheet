# K-Means & DBSCAN clustering
This notebook demonstrates usage of K-Means and DBSCAN clustering algorithms. Both approaches are demonstrated on two types of data: blob and moon. Key takeaway from this notebook is that you should pick models with respect to characteristics of a data.

### Imports

```python
import numpy as np
from sklearn.cluster import KMeans, DBSCAN
from sklearn.datasets import make_blobs, make_moons
from matplotlib import pyplot as plt
```

## Blob data

### Generate data

```python
data, labels = make_blobs(n_samples=200, random_state=1)
```

```python
plt.scatter(data[:, 0], data[:, 1], c=labels)
plt.show()
```

### Create & apply KMeans clustering

```python
km = KMeans(n_clusters=3)
km.fit(data)

plt.scatter(data[:, 0], data[:, 1], c=km.labels_)
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], marker="x", c="red")
plt.show()
```

### Create & apply DBSCAN clustering
Input parameters:
- `eps` – Controls local neighborhood of a point. Crutial to set up correctly. Small `eps` -> tendency to set everything as "noise" (label == -1), high `eps` -> tendency to create one huge cluster.
- `min_samples` - Tolerancy towards noise. Large data with a lot of noise -> higher `min_samples`. 

```python
db = DBSCAN(eps=1.0, min_samples=3)
db.fit(data)

plt.scatter(data[:, 0], data[:, 1], c=db.labels_)
plt.show()
```

## Moon data

### Generate data

```python
data, labels = make_moons(n_samples=200, noise=0.1, random_state=1)
```

```python
plt.scatter(data[:, 0], data[:, 1], c=labels)
plt.show()
```

### Create & apply KMeans clustering

```python
km = KMeans(n_clusters=2)
km.fit(data)

plt.scatter(data[:, 0], data[:, 1], c=km.labels_)
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], marker="x", c="red")
plt.show()
```

### Create & apply DBSCAN clustering


```python
db = DBSCAN(eps=0.25, min_samples=8)
db.fit(data)

plt.scatter(data[:, 0], data[:, 1], c=db.labels_)
plt.show()
```
